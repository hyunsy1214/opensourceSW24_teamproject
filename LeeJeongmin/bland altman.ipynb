```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.lines as mlines
# if using a Jupyter notebook, include:
%matplotlib inline

import pandas as pd
import statistics as stat
```

```python
data1 = [44.3,33.3,25,100,90,80,20,50,65]
data2 = [55.2,44.3,30,88,88,77,22,66,70]
mean_of_two = [49.75,38.8,27.5,94,89,78.5,21,58,67.5]
diff = [-10.9,-11,-5,12,2,3,-2,-16,-5]
```

```python
mean = np.mean(diff)
std = np.std(diff)
var = np.var(diff)
upper = mean + (std*1.96)
lower = mean - (std*1.96)

print ("\nmean: ", mean)
print ("\nstandard deviation: ", std)
print ("\nvariance: ", var)
print ("\n")
print ("\nupperlimit: ", upper)
print ("\nlowerlimit: ", lower)
```

```python
#수식으로 표준편차, 분산 구하기

mean2 = sum(diff)/len(diff)
print("\nmean: ", mean2)

std2 = np.sqrt(np.mean((diff - np.mean(diff)) ** 2)) 
print("\nstandard deviation: ", std2) 
  
var3= np.mean((diff - np.mean(diff)) ** 2) 
print("\nvariance: ", var3) 
```

```python
plt.figure(figsize=(10,7))

plt.scatter(mean_of_two, diff, color ='salmon', linewidth =2)
plt.xlabel('Average of two Segmentation', fontsize = 15)
plt.ylabel('Difference of two segmentation', fontsize = 15)
plt.xlim(20,100)
ax = plt.subplot()
ax.set(xticks =[40,60,80,100],yticks =[20,15,10,5,0,-5,-10,-15,-20,-25])

# 그리드라인
ax.grid(False) # 없애고 싶으면 False 로 바꾸기

ax.spines['right'].set_visible(False)
ax.spines['top'].set_color('white')

# 추가선 그리기
mean1 = [20, mean]
mean2 = [100, mean]
upper3 =[20, upper]
upper4 =[100, upper]
lower5 =[20, lower]
lower6 =[100, lower]

x = [mean1[0], mean2[0]]
y = [mean1[1], mean2[1]]
plt.plot(x,y)

x1 = [upper3[0], upper4[0]]
y1 = [upper3[1], upper4[1]]
plt.plot(x1,y1, '--', c='salmon', linewidth=2)

x2 = [lower5[0], lower6[0]]
y2 = [lower5[1], lower6[1]]
plt.plot(x2,y2, '--', c='salmon', linewidth=2)

```
